```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
city_data.head()

ride_data = pd.read_csv(ride_data_to_load)
ride_data.head()

# Combine the data into a single dataset
city_ride_df= pd.merge(ride_data, city_data, on ='city', how='left')


# Display the data table for preview
city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Calculating avg fare, total rides & driver count for Urban city type
avgfare_urban = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Urban')].groupby(['city'])['fare'].mean())
totalrides_urban = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Urban')].groupby(['city'])['ride_id'].count())
drivers_urban = city_ride_df.loc[(city_ride_df['type'] == 'Urban')][['city', 'driver_count']].drop_duplicates().set_index('city')

#Calculating avg fare, total rides & driver count for Suburban city type
avgfare_sub = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Suburban')].groupby(['city'])['fare'].mean())
totalrides_sub = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Suburban')].groupby(['city'])['ride_id'].count())
drivers_sub = city_ride_df.loc[(city_ride_df['type'] == 'Suburban')][['city', 'driver_count']].drop_duplicates().set_index('city')

#Calculating avg fare, total rides & driver count for Rural city type
avgfare_rural = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Rural')].groupby(['city'])['fare'].mean())
totalrides_rural = pd.DataFrame(city_ride_df.loc[(city_ride_df['type'] == 'Rural')].groupby(['city'])['ride_id'].count())
drivers_rural = city_ride_df.loc[(city_ride_df['type'] == 'Rural')][['city', 'driver_count']].drop_duplicates().set_index('city')


#Defining the size of the graph
plt.figure(figsize= (10,8))

#Building the scatter plots for each city types
plt.scatter(totalrides_urban, avgfare_urban, marker='o', s=drivers_urban*15, facecolors='lightcoral', edgecolors ='black', label = 'Urban', alpha = 0.6)
plt.scatter(totalrides_sub, avgfare_sub, marker='o', s=drivers_sub*15, facecolors='lightskyblue', edgecolors ='black', label = 'Suburban', alpha = 0.6)
plt.scatter(totalrides_rural, avgfare_rural, marker='o', s=drivers_rural*15, facecolors='gold', edgecolors ='black', label = 'Rural', alpha = 0.6)

#Adding some other graph properties like grid, title, x lable & y label
plt.grid()
plt.suptitle('Pyber Ride Sharing Data (2016)')
plt.title('Note: Circle size correlates with driver count per city')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare($)')

#Inserting a legend and determining the size for each city type
lgd = plt.legend(loc='best', edgecolor = 'black', title = 'City Types')
lgd.legendHandles[0]._sizes = [60]
lgd.legendHandles[1]._sizes = [60]
lgd.legendHandles[2]._sizes = [60]
# plt.figure(figsize= (15,10))


#Save and show the scatter plot
plt.savefig('Pyber_Ride_Sharing.png')
plt.show()
```

The following conclusions can be made by the above scatter plot:

1. The Urban area shows the highest number of ride sharing, whereas suburban areas come second. As we can see the Rural areas have minimal ride sharing as compared to Urban & Suburban.

2. Average fares are less in the urban areas. Then it increases in Suburban but the Rural has the highest average fare.

3. The number of drivers are also high in Urban areas which is because of the high demand. Suburban area also has good number of drivers comparatively. And again the driver supply decreases in Rural areas.

## Total Fares by City Type

```python
#Declaring a color list to be used for next three pie charts
colors_list = ['gold', 'skyblue', 'lightcoral']

#Declaring explode for next three pie charts
explode = (0, 0, 0.1)

#Grouping by Type for next three pie charts
type_grp = city_ride_df.groupby('type')['type', 'ride_id', 'fare', 'driver_count']
type_grp.head()
```

```python
#Calculating the total fare by using above delared group by type
totalfare_type = type_grp.sum()['fare']

#Creating a label list using the index of the calculatedn fare
labels_list = totalfare_type.index

#Defining the size of the graph
plt.figure(figsize= (10,8))

#Plotting the pie chart and adding title
plt.pie(totalfare_type, explode=explode, labels=labels_list, colors=colors_list, autopct="%1.1f%%", 
        shadow=True, startangle=160,textprops={'fontsize':12})
plt.title('% of Total Fare by City Type')


#Save and show the pie chart
plt.savefig('Total_Fares_By_City_Type.png')
plt.show()
```

## Total Rides by City Type

```python
#Calculating the total rides by using above delared group by type
totalride_type = type_grp.count()['ride_id']


#Creating a label list using the index of the calculatedn fare
labels_list = totalride_type.index

plt.figure(figsize = (12,10))

#Defining the size of the graph
plt.figure(figsize= (10,8))

#Plotting the pie chart and adding title
plt.pie(totalride_type, explode=explode, labels=labels_list, colors=colors_list, autopct="%1.1f%%", 
        shadow=True, startangle=160, textprops={'fontsize':12})
plt.title('% of Total Rides by City Type')


#Save and show the pie chart

plt.savefig('Total_Rides_By_City_Type.png')
plt.show()
```

## Total Drivers by City Type

```python
#Calculating the total drivers by using above delared group by type
totaldriver_type = type_grp.sum()['driver_count']


#Creating a label list using the index of the calculatedn fare
labels_list = totalride_type.index

#Defining the size of the graph
plt.figure(figsize= (10,8))

#Plotting the pie chart and adding title
plt.pie(totaldriver_type, explode=explode, labels=labels_list, colors=colors_list, autopct="%1.1f%%", 
        shadow=True, startangle=160,textprops={'fontsize':12})
plt.title('% of Total Drivers by City Type')


#Save and show the pie chart

plt.savefig('Total_Drivers_By_City_Type.png')
plt.show()

```

After looking at all the three pie charts, following conclusions can be made on the returns:

1. The Urban areas provide considerable revenue of 62.7% which is way higher than Suburban (30.5%) and Rural (6.8%).
2. The total number of drivers are very high in Urban areas which explains why it returns the most revenue. Even though the average fare is less in Urban, the total rides in those areas increased the revenue.
3. There could be possible growth in Suburban if provided more drivers, which could possibly decrease the fare and will attract more people there too.
